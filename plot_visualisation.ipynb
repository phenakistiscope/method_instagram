```python
#import librairies
import pandas
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
#import data from your directory
df = pandas.read_csv("path_to/file.csv")
```

Step 2. Change country name form two letters to full name, switch date column to datetime type and clean rows in  data frame

```python
#import iso.csv file who is in repository
df_iso = pandas.read_csv("path_to/iso.csv")
```

```python
dff = pandas.merge(df_iso, df, on='country',how='outer')
dff.dropna(subset=['country'], inplace=True)
dff.dropna(subset=['date'], inplace=True)
dff['date'] = pandas.to_datetime(dff['date'])
dff['year'] = pandas.DatetimeIndex(dff.date).year
dff['month'] = pandas.DatetimeIndex(dff.date).month
dff['day'] = pandas.DatetimeIndex(dff.date).day
```

Step 3. plot data 

This first plot is made for understanding the frequency of publication in a country related to all the publications. The choice is to get only countries one by one otherwise the diagram is unclear. 

```python
def plot_year_country(year: int, country: str, dataframe):
    data = dataframe
    data1 = data[(data['year'] == year)]    
    data2 = data1[(data1['name'] == country)]

    fig = plt.figure(figsize=(14,10))
    ax1 = fig.add_subplot(1,1,1)
    data1.date.value_counts().plot(color="#663399",alpha=0.6, label = 'World')
    data2.date.value_counts().plot(color='r',alpha=1, label = country)
    ax1.legend(loc='best')
    plt.title("Publications by day in " +(country) +" for the year " +(str(year)), fontsize=20)
    ax1.set_ylabel('Publications')
    ax1.xaxis.set_major_locator(ticker.AutoLocator())
    return fig
```

```python
plot_year_country(year=2019, country='Switzerland', dataframe = dff) #.savefig('plot_Country_year.png')
```

This third plot is more precise. It helps us to see the frequency in a country and also the likes who needs to be precisly attached to a publication or a cluster of them. In this sens this plot goes with the last stetp.

```python
def plot_like(year: int, country: str, dataframe):
    dff = dataframe
    dff1 = dff[(dff['year'] == year)]
    dff2 = dff1[(dff1['name'] == country)]

    fig = plt.figure(figsize=(14,10))
    ax1 = fig.add_subplot(1,1,1)
    ax1.hist(dff2.date, bins=80,color='b',alpha=0.6, histtype="barstacked", label = (str(country)))
    ax2 = ax1.twinx()
    ax2.scatter(dff2.date, dff2.likes, color='r',alpha=0.8, label='Likes')
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
    plt.title('Publications and likes by day in ' + (str(country)) + ' for the year ' +(str(year)), fontsize=20)
    ax1.set_ylabel('Publications')
    ax2.set_ylabel('Likes')
    ax2.yaxis.set_major_locator(ticker.MaxNLocator())
    return fig
```

```python
plot_like(year=2019, country='Hong Kong', dataframe= dff)#.savefig('plot_likes_HongKong_2019_img.png')
```

Step 4. This step represents the transition from quantitative to qualitative. The following fonction allows us to search picture depanding what we've learn before. By plots functions we are able to understand different scale and this last step lead us back to our visual expertise image by image.

```python
def images(year: int, month: int, country: str, dataframe):
    dff= dataframe
    dff1 = dff[(dff['year'] == year)]
    dff2 = dff1[(dff1['month'] == month)]
    dff3 = dff2[(dff2['name'] == country)]
    return dff3[['id', 'city_one', 'day']]
```

```python
list_img = images(year=2019, month=2, country= 'Germany', dataframe=dff)
list_img.to_csv('liste_images_Germany_2019_2.txt', header=True, index=False, sep='\t', mode='a')
```
